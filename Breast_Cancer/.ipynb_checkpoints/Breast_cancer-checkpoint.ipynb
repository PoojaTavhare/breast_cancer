```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("BreastCancer.csv")
```

```python
df.tail()
```

```python
#df["diagnosis"]= df["diagnosis"].map({"M": 1, "B": 0})
```

```python
X=df.drop(["diagnosis","Unnamed: 32","id"],axis=1)
Y= df[["diagnosis"]]
```

```python
X
```

```python
Y
```

```python
df["diagnosis"].value_counts(normalize = True)
```

```python
    from sklearn.model_selection import train_test_split
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.70, random_state = 45, stratify = df["diagnosis"])
    

    from sklearn.preprocessing import MinMaxScaler
    scalling = MinMaxScaler()
    X_train_sc = scalling.fit_transform(X_train)   
    X_test_sc = scalling.transform(X_test)   


    from sklearn.linear_model import LogisticRegression
    log_reg = LogisticRegression()  
    log_reg.fit(X_train_sc,Y_train)
    Y_pred = log_reg.predict(X_test_sc)   
    
```

```python
log_reg.coef_     
```

```python
log_reg.intercept_ 
```

```python
from sklearn import metrics
metrics.classification_report(Y_test, Y_pred)   
conf_mat = metrics.confusion_matrix(Y_test, Y_pred)   
    
import matplotlib.pyplot as plt
disp = metrics.ConfusionMatrixDisplay(conf_mat)   
disp.plot()                                                                              
plt.show()
```

```python
metrics.fbeta_score(Y_test, Y_pred, beta = 1)   
metrics.fbeta_score(Y_test, Y_pred, beta = 0.5)
metrics.fbeta_score(Y_test, Y_pred, beta = 2 )   
            
fpr, tpr, threshold = metrics.roc_curve(Y_test, Y_pred)
```

```python
metrics.fbeta_score(Y_test, Y_pred, beta = 1)
```

```python
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], "--r", label = "Base Line")
plt.xlabel("False Positive Rate (1 - Specificity)")
plt.ylabel("True Positive Rate (Sensitivity)")
plt.title("ROC Curve")
plt.legend()
plt.show()
metrics.roc_auc_score(Y_test, Y_pred)
```
